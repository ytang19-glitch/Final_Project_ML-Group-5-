# **Introduction**

Air pollution has emerged as one of the most critical environmental and public health concerns in rapidly developing nations, and India stands at the forefront of this challenge. With rising industrialization, urban expansion, vehicular emissions, and seasonal agricultural practices, air quality in many Indian cities frequently exceeds safe limits. The Air Quality Index (AQI) has become a widely adopted metric for summarizing pollution levels and their associated health risks, enabling policymakers, health agencies, and citizens to respond appropriately to hazardous conditions. However, despite the availability of AQI reporting systems, most solutions are reactive in nature, informing the public only after pollution levels have deteriorated.

To address this gap, data-driven approaches and machine learning techniques offer valuable opportunities for proactive environmental monitoring. By uncovering temporal patterns, pollutant correlations, and seasonal influences through Exploratory Data Analysis (EDA), we gain critical insights into how factors such as particulate matter, gas emissions, and meteorological conditions shape air quality trends. These insights highlight not only the complexity of air pollution but also the potential for intelligent forecasting systems that leverage historical data to anticipate future conditions rather than simply describe the present.

The primary objective of this project is therefore to develop a machine learning model capable of predicting the Air Quality Index (AQI) three days in advance. By enabling short-term AQI forecasting, the system aims to support early warnings, strategic planning, and informed decision-making for both the public and government bodies. Such a predictive framework has the potential to mitigate exposure to harmful pollutants, optimize healthcare preparedness, and guide interventions that improve urban air quality. Ultimately, this research demonstrates how AI-enabled environmental analytics can transition air quality management from reactive reporting to proactive prevention.

# **Methodology**

The methodology employed in this project follows a systematic data-driven framework composed of five major phases: data acquisition, preprocessing, exploratory data analysis, model development, and model evaluation. Each phase is designed to progressively transform raw environmental data into meaningful insights and actionable AQI predictions.

1. Data Acquisition
Historical air quality data for selected Indian cities was obtained from publicly available datasets. The dataset includes daily AQI values along with concentrations of various pollutants such as PM2.5, PM10, CO, NO, NO₂, SO₂, O₃, and NH₃. Temporal attributes, including date, month, and year, were also available and were used to derive seasonal information relevant to pollution patterns.

2. Data Preprocessing
The dataset was cleaned and organized to ensure compatibility with machine learning workflows. This involved handling missing values, removing irrelevant features, standardizing date formats, converting string-based dates into datetime objects, and extracting new temporal variables such as month, season, and day of year. The data was then sorted and grouped by city to allow city-specific analysis. Only numerical attributes were retained for model training, and categorical columns were encoded where necessary.

3. Exploratory Data Analysis (EDA)
EDA was conducted to identify underlying relationships and trends in the dataset. Statistical summaries, correlation heatmaps, distribution plots, time-series graphs, and seasonal analysis were used to examine how pollutant concentrations and weather cycles influence AQI. This step helped uncover seasonal patterns—such as elevated AQI during winter months and decreased levels during the monsoon season—and provided justification for including temporal features in the predictive model.

4. Model Development
Machine learning models were developed to forecast AQI values three days ahead using historical pollutant levels and temporal features. The dataset was split into training and testing subsets, and regression-based algorithms were implemented. Independent variables included pollutant concentrations and season-based temporal indicators, while AQI served as the target output. Feature selection techniques were applied to ensure that only relevant predictors were included in the model.

5. Model Evaluation
Model performance was assessed using appropriate metrics to measure prediction accuracy and reliability. The forecasted AQI values were compared to actual observations, and the error distributions were analyzed to understand model strengths and limitations. The final evaluation determined the model’s suitability for short-term AQI forecasting and identified areas for future enhancement.

AQI Scoring:
1 - Good (0 - 50)
2 - Moderate (51 - 100)
3 - Unhealthy for selected groups (101 - 150)
4 - Unhealthy (151 - 200)
5 - Very Unhealthy (201 - 300)

*The following pollutant concentrations are measured in μg/m³:*

**CO:** Carbon Monoxide, a colourless and odourless gas released from incomplete combustion of fuels and biomass.

**NO:** Nitric Oxide, a pollutant generated mainly from fuel combustion and industrial activities.

**NO₂:** Nitrogen Dioxide, a toxic gas largely produced by vehicle emissions and combustion processes.

**O₃:** Ozone, a secondary pollutant formed when sunlight interacts with nitrogen oxides and volatile organic compounds.

**SO₂:** Sulphur Dioxide, a gas emitted from burning fossil fuels, especially in power generation and heavy industries.

**PM2.5:** Fine particulate matter with a diameter of 2.5 micrometres or smaller, capable of penetrating deep into the lungs.

**PM10:** Coarse particulate matter with a diameter of 10 micrometres or smaller, originating from dust, smoke, and other airborne particles.

**NH₃:** Ammonia, a gas produced mainly from agricultural activities, livestock waste, and some industrial operations.

# **Initial Analysis of Datasets**




**1. Correlation Heatmap**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('/content/air_pollution_data.csv')   # or whatever your file is

sns.set_theme()  # <- nice default styling

numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Air Quality Variables', fontsize=14)
plt.show()
```

**Interpretation:** The correlation heatmap shows clear relationships among air quality variables. AQI exhibits the strongest positive correlations with PM10 (0.47) and PM2.5 (0.46), confirming that particulate matter is the primary driver of poor air quality in the dataset. Carbon monoxide (0.38) and nitrogen dioxide (0.44) also show moderate correlations with AQI, indicating they contribute to pollution levels but to a lesser extent.

Among the pollutants, PM2.5 and PM10 are almost perfectly correlated (0.99), meaning both particles tend to increase together and represent the same pollution source profile—likely vehicular and industrial emissions. CO, NO, and NO2 also show strong inter-correlations (0.72–0.94), suggesting they originate from similar combustion activities.

In contrast, ozone (O3) displays weak or negative correlations with most pollutants and AQI, implying that it behaves differently from particulate and combustion pollutants, often influenced by sunlight-driven chemical reactions rather than direct emissions.

Overall, the heatmap confirms that particulate matter (PM2.5 and PM10) is the dominant factor affecting AQI, while combustion gases (CO, NO, NO2) act as secondary contributors.

## 5. Monthly AQI Trend ##

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Robust datetime parsing
df['date'] = pd.to_datetime(
    df['date'],
    format='mixed',
    dayfirst=True,
    errors='coerce'
)

# 2. Drop rows where date could not be parsed (optional but cleaner)
df = df.dropna(subset=['date'])

# 3. Extract month names
df['month'] = df['date'].dt.month_name()

# 4. Compute average AQI per month
monthly_aqi = df.groupby('month')['aqi'].mean()

# 5. Reorder months in calendar order
order = ['January','February','March','April','May','June','July',
         'August','September','October','November','December']
monthly_aqi = monthly_aqi.reindex(order)

# 6. Plot
plt.figure(figsize=(12,6))
sns.lineplot(x=monthly_aqi.index, y=monthly_aqi.values, marker='o', linewidth=2)

plt.title('Monthly Trend of Average AQI', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Average AQI')
plt.xticks(rotation=45)
plt.grid(alpha=0.3, linestyle='--')
plt.show()
```

## Interpretation ##

The monthly AQI pattern, where pollution peaks between December and March, is consistent with the climate and emission characteristics of many cities in your dataset. Several included cities—such as Delhi, Lucknow, Patna, Jaipur, and Kolkata—are located in the Indo-Gangetic Plain, a region well-documented for severe winter smog. During winter, low temperatures and stagnant air prevent pollutant dispersal, causing particulate matter and combustion gases to accumulate close to the ground.

Cities like Delhi and Lucknow are especially affected by:

- Temperature inversion in winter, which traps pollutants

- Post-harvest stubble burning in surrounding states

- High traffic and industrial emissions

This explains why these cities show consistently high AQI values and contribute to the elevated averages during winter months.

In contrast, southern and coastal cities in your dataset such as Kochi, Thiruvananthapuram, Visakhapatnam, and Coimbatore experience stronger monsoon winds and higher humidity, which naturally disperse pollutants. These cities typically exhibit lower AQI values and greater seasonal improvement during monsoon periods, aligning with the lowest AQI seen in June to August on your chart.

Northeastern cities like Shillong and Guwahati also have cleaner air due to lower industrial density and more favorable atmospheric conditions, reinforcing the dip in AQI during mid-year.

## 6. Seasonal AQI Variation by City Group ##

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure date is already parsed (from previous steps)
# df['date'] = pd.to_datetime(df['date'], format='mixed', dayfirst=True, errors='coerce')

# Extract month and assign season
df['month'] = df['date'].dt.month

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'

df['season'] = df['month'].apply(get_season)

# Plot
plt.figure(figsize=(10,6))
sns.boxplot(x='season', y='aqi', data=df, order=['Winter','Summer','Monsoon','Post-Monsoon'])

plt.title('Seasonal Variation of AQI', fontsize=16)
plt.xlabel('Season')
plt.ylabel('AQI')
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.show()
```

## Explanation

This density contour plot illustrates how Air Quality Index (AQI) levels vary throughout the year across Indian cities. The x-axis represents the months, while the y-axis shows the AQI values. The colored regions show where AQI observations are most concentrated—brighter colors (yellow/orange) indicate a high density of data points, meaning those AQI levels were frequently recorded during that month; darker colors (purple/blue) indicate fewer occurrences.

A clear seasonal trend can be observed:

- Winter months (December–February) show the highest density at AQI levels around 4–5, indicating frequent poor air quality. This aligns with India’s winter weather patterns, where stagnant air, temperature inversion, crop-burning practices (especially in northern states), and increased heating contribute to pollution buildup.

- Spring and early summer (March–May) show a gradual decline in AQI density, moving toward moderate levels. Windier conditions and rising temperatures help disperse pollutants.

- Monsoon months (June–August) display the lowest AQI densities, clustered around AQI 2–3, indicating significantly improved air quality. Rainfall naturally cleans the atmosphere by removing particulate matter.

- Post-monsoon and early winter (September–November) show AQI rising again, reflecting agricultural burning, festival fireworks (such as Diwali in October/November), and reduced atmospheric movement.

Overall, the plot confirms a strong seasonal cycle: air quality improves dramatically during the monsoon and deteriorates sharply during winter. This reinforces the importance of seasonal forecasting in air quality management and justifies using time-aware machine learning models for AQI prediction.
